```python
import urllib.request as urllib2
from bs4 import BeautifulSoup
from datetime import datetime
import re
```

```python
def getArticleText(url):
    soup = BeautifulSoup(urllib2.urlopen(url))
    result = soup.find("div", {"class": "entry-content"})
    for div in result.findAll("div"):
        div.decompose()
    return result.text

def getArticleDate(url):
    soup = BeautifulSoup(urllib2.urlopen(url))
    return soup.find("div", {"id": "left_element_header"}).text

def getArticlesUrl(main_url):
    article_urls = []
    soup_all = BeautifulSoup(urllib2.urlopen(main_url + "/waluty/wiadomosci"))
    articles_divs = soup_all.findAll("div", {"class": "article"})
    for div in articles_divs:
        soup_content = div.find("div", {"class": "entry-content"})
        span = soup_content.find("span", {"class": "entry-title"})
        href = span.find("a")["href"]
        article_urls.append(main_url + href)
    return article_urls

def getArticleTextDateTuples(article_urls):
    articles = []
    for url in article_urls:
        print(url)
        text, date = getArticleText(url), getArticleDate(url)
        text = text.replace('\r', '').replace('\n', '')
        date = re.search('[0-9]+[-][0-9]+[-][0-9]+', date).group()
        date = str(datetime.strptime(date.replace("\n", ""), "%Y-%m-%d").date())
        articles.append(tuple((text, date)))
    return articles
```

```python
main_url = "https://www.bankier.pl"
article_urls = getArticlesUrl(main_url)

print(getArticleTextDateTuples(article_urls))
```
